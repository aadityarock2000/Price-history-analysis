```python
# Create table
# for each commodity:
#     find pre coid rate
#     find post covid rate
#     sort by post covind rate
```

```python
pre_cov,post_cov,commodity_table_list=[],[],[]
for i in range(1,23):
    commodity=column_list[i]
    df_month=df.groupby([pd.PeriodIndex(df['Date'], freq="Y"),'City'])[commodity].mean()
    df_month=pd.DataFrame(df_month)
    df_month=df_month.reset_index()
    temp=df_month[df_month['City']=='LUCKNOW']
    temp['diff']=temp[commodity].diff()
    
    
    pre_cov.append(100*(temp.iloc[5,2]-temp.iloc[0,2])/temp.iloc[0,2])
    
    post_cov.append(100*(temp.iloc[8,2]-temp.iloc[6,2])/temp.iloc[6,2])
    
    commodity_table_list.append(commodity)

temp_inflation_rates=pd.DataFrame()
temp_inflation_rates['pre_cov']=pre_cov
temp_inflation_rates['post_cov']=post_cov
temp_inflation_rates['diff']=temp_inflation_rates['post_cov']-temp_inflation_rates['pre_cov']
temp_inflation_rates['commodity']=commodity_table_list
    
    
    
```

```python
for i in range(1,23):
    commodity=column_list[i]
    df_yearly=df.groupby([pd.PeriodIndex(df['Date'], freq="Y"),'City'])[commodity].mean()
    df_yearly=pd.DataFrame(df_yearly)
    df_yearly=df_yearly.reset_index()

    #finding the minimum and maximum
    result1 = df_yearly.groupby('Date').min()
    result1.drop('City',inplace=True,axis=1)
    result2 = df_yearly.groupby('Date').max()
    result2.drop('City',inplace=True,axis=1)

    #inner join on the date for the max and min
    result=result2.merge(result1,how='inner',on='Date')

    #finding the max and min cities and the percentage
    minimum=commodity+'_y'
    maximum=commodity+'_x'
    result['Difference']=result[maximum]-result[minimum]
    result['Difference percentage']=100*(result['Difference']/result[minimum])
    yearly_diff[commodity]=result['Difference percentage']
    
```
